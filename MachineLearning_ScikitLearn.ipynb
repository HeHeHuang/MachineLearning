### PCA



```python
import pandas as pd 
import numpy as np
import random as rd
from sklearn.decomposition import PCA
from sklearn import preprocessing
import matplotlib.pyplot as plt

```

```python
genes = ['gene'+str(i) for i in range(1,101)]
```

```python
genes
```

```python
wt = ['wt' +str(i) for i in range(1,6)]
ko = ['ko'+str(i) for i in range(1,6)]
```

```python
data = pd.DataFrame(columns = [*wt,*ko],index = genes)
data
```

```python
for gene in data.index:
  data.loc[gene,'wt1':'wt5'] = np.random.poisson(lam=rd.randrange(10,1000),size =5)
  data.loc[gene,'ko1':'ko5'] = np.random.poisson(lam=rd.randrange(10,1000),size =5)
```

```python
print(data.head())
```

```python
print(data.shape)
```

```python
 scaled_data = preprocessing.scale(data.T)   # after centering, the average value for each gene will be 0, and after scaling, the standard deviation for the values for each gene will be 1 
 # variation is calculated as (measurements- mean) square / the number of measurements 
 scaled_data
```

```python
 pca = PCA()
 pca.fit(scaled_data)
 pca_data = pca.transform(scaled_data)
```
